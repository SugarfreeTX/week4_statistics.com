```python
# We only need date, month and year 
# A count of how many released
# Average profitability for those movies. 
```

```python
def make_date(date_str):
    m, d, y = date_str.split('/')
    m = int(m)
    d = int(d)
    y = int(y)
    if y > 13:
        y += 1900
    else:
        y += 2000
    return dt.datetime(y, m, d)
```

```python
movies = pd.read_csv("MovieData.csv", sep='\t', 
                         na_values=["Unknown", "Unkno"], parse_dates=[0], date_parser=make_date)
```

```python
movies.head()
```

```python


def get_year(date):
    return date.year

def get_month(date):
    return date.month 
```

```python
movies["Month"] = movies.Release_Date.apply(lambda x: x.month)
movies["Year"] = movies.Release_Date.apply(lambda x: x.year)
```

```python
movies.fillna(0, inplace=True)
```

```python
movies["Worldwide Gross"][movies["Worldwide Gross"]==0] = movies["US Gross"]
```

```python

movies["Profit"] = movies["Worldwide Gross"] -  movies["Budget"]
```
