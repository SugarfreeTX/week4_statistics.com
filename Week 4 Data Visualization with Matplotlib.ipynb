## Figures and subplots:

```python
# Generate new data. 
x_data = range(100)
y_data = [np.random.random() for x in x_data]
```

```python
# Testing what plot does without creating a figure. 
fig_1 = plt.figure()
ax1 = fig_1.add_subplot(111)
ax1.plot(x_data, y_data)
```

```python
fig = plt.figure() # Create the new figure.
ax = fig.add_subplot(111) # Create a single subplot within the figure.
ax.plot(x_data, y_data) # Plot the data in the subplot
```

```python
# the plt.fig() function creates a new figure. 
# add_subplot() method creates a new subplot within that figure that is 
# assigned to variable ax. 
```

```python
# The number in .add_subplot() it's purpose is to tell matplotlib 
# how to divide a figure into subplots.
# The first (hundred place) is number of rows. 
# The second is number of columns.
# The third is the index of the specific subplot to plot in. 
# Count across and then down. 
```

```python
fig = plt.figure(figsize=(5,5)) # Set the figure size to be 5x5
ax = fig.add_subplot(111)
ax.plot(x_data, y_data)
```

## Manipulating figures and axes:

```python
# Some more random data. 
x_data = np.linspace(0, 10, 21)
y_data = [np.random.randint(0, 10) for x in x_data] # Now we pick a random 
# integer between 0 and 10.
```

Three most often used: 
* set_title() give the subplot a heading
* set_xlabel() give a label to the x axis 
* set_ylabel() give a label to the y axis 

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)

ax.set_title("Some random numbers")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.plot(x_data, y_data)
```

```python
# Using set_xlim() and .set_ylim()
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)

ax.set_title("Some random numbers")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.plot(x_data, y_data)

ax.set_xlim(-10, 20) # row
ax.set_ylim(-1 , 10) # column 
```

```python
# Using set_xlim() and .set_ylim()
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)

ax.set_title("Some random numbers")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.plot(x_data, y_data)

ax.set_ylim(5, 10)
```

## Styling plots 

```python
# Generate two sets of data, one x axis with two sets of y points. 
x_data = range(20)
y1_data = [np.random.randint(0, 10) for x in x_data]
y2_data = [np.random.randint(0, 10) for x in x_data]
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
ax.plot(x_data, y1_data)
ax.plot(x_data, y2_data)
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
ax.plot(x_data, y1_data, c='k') # black
ax.plot(x_data, y2_data, c='r') # red
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
ax.plot(x_data, y1_data, c='k', linewidth=2.0) # black # linewidth adjusted
ax.plot(x_data, y2_data, c='r', linestyle='--') # red # linestyle dotted
```

```python
# Can mix any kind of chart. Overlay.
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
ax.plot(x_data, y1_data, c='k')
ax.scatter(x_data, y2_data, c='r')
ax.grid(True) # Turn on the grid lines. 
```

```python
# The attribute s for scatter plots allows you to change the size of points.
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
ax.scatter(x_data, y1_data, s=y2_data)
```

```python
# Passing a list of numbers for s= will add dimension to the visual.
# To scale the dimensions for better size value. Multiply by 16 to get size. 
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
ax.scatter(x_data, y1_data, s=[16 * y for y in y2_data])
```

```python
# To plot color and remove outline. 
# Plot points with y2 below 5 in red
# and points above 5 in green. 
colors = []
for y in y2_data:
    if y < 5: colors.append('r')
    else: colors.append('g')
        
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
ax.scatter(x_data, y1_data, s=32, c=colors, linewidths=0)
```

## Legends

```python
x_data = range(20)
y1_data = [np.random.randint(0, 10) for x in x_data]
y2_data = [np.random.randint(0, 10) for x in x_data]
```

```python
# Just to plot. 
plt.plot(x_data, y1_data)
plt.plot(x_data, y2_data)
plt.legend(["Line 1", "Line 2"])
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
ax.plot(x_data, y1_data, c='k', linewidth=2.0)
ax.plot(x_data, y2_data, c='r', linestyle='--')

ax.legend(["Black line", "Dashed red line"], loc='lower left')
```

```python
# Moving the legend outside the subplot by using number 1 or greater.
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
ax.plot(x_data, y1_data, c='k', linewidth=2.0)
ax.plot(x_data, y2_data, c='r', linestyle='--')

ax.legend(["Black line", "Dashed red line"], loc=(1.05, 0))
```

```python
# Phantom points used for scatter plot legends. 
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
ax.scatter(x_data, y1_data, s=[16 * y for y in y2_data])

# Creating proxy artists
#p4 = ax.scatter([-10], [0], s=16, c = 'C0')
p1 = ax.scatter([], [], s=16, c='C0')
p2 = ax.scatter([], [], s=16*5, c='C0')
p3 = ax.scatter([], [], s=16*10, c='C0')

# Plotting the legend
ax.legend([p1, p2, p3], ["1", "5", "10"], loc=(1.01, 0))
```

# Histograms and binning

```python
data = [2.4, 4.71, 2.53, 3.45, 1.23, 1.53, 2.37, 3.35, 2.67, 3.16, 
1.31, 4.78, 4.44, 1.64, 1.13, 2.68, 4.76, 1.79, 2.89, 3.58, 4.1, 
4.81, 1.14, 4.09, 4.99, 2.47, 4.77, 1.09, 3.56, 1.26, 3.26, 4.56, 
1.81, 4.36, 1.83, 3.68, 4.21, 1.15, 4.85, 1.17]
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)

hist = ax.hist(data)
```

```python
# To assign bins ourselves we need to assign LOV to the bins property. 
# The numbers indicate the start and end points. 
# By default the hist will give us counts - the number of items in each bin. 
# We can change to relative freq (the fraction of items in each bin) 
# by setting normed = True
fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(111)

hist = ax.hist(data, bins=range(1,6), normed=True)
```

```python
fig = plt.figure(figsize = (6,4))
ax = fig.add_subplot(111)
ax.set_ylim(0, 15)

hist = ax.hist(data, bins=np.linspace(0, 6, 5), rwidth=0.8, color='g')
```

# Multiple subplots

```python
# First set random numbers
# Second set difference between those numbers 
# y2_data[i] = y1_data[i] = y1_data[i-1]
# y2 will be positive when y1 increases and negative when it decreases. 
# Keep aligned, if y1 counts from 0. y2 counts from 1. 
```

```python
x1_data = range(10)
y1_data = [np.random.random() for x in range(10)]

x2_data = range(1, 10)
y2_data = [(y1_data[i] - y1_data[i-1]) for i in range(1, 10)]
```

```python
# Create one figure:
fig = plt.figure(figsize =(12, 6))

ax1 = fig.add_subplot(121) # Create the first subplot
ax1.plot(x1_data, y1_data) # and plot the data in it

ax2 = fig.add_subplot(122) # Now create the second subplot
ax2.plot(x2_data, y2_data) # and plot the data there too. 
```

```python
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(211)
ax1.plot(x1_data, y1_data)

ax2 = fig.add_subplot(212)
ax2.plot(x2_data, y2_data)
ax2.set_xlim(0, 9) # Keep them aligned
```

```python
fig = plt.figure(figsize=(6, 6))

ax1 = fig.add_subplot(221)
ax1.plot(x1_data, y1_data)

ax2 = fig.add_subplot(222)
ax2.hist(y1_data)

ax3 = fig.add_subplot(223)
ax3.plot(x2_data, y2_data)
ax3.set_xlim(0, 9)

ax4 = fig.add_subplot(224)
ax4.hist(y2_data, color='g')
```

# Plotting pandas data

```python
import pandas as pd
```

```python
data = pd.read_csv("GoldAndOil.csv", parse_dates=[0]) # Load the data. 
```

```python
data.head()
```

```python
data.set_index("DATE", inplace=True)
```

```python
data.head()
```

```python
# Call dataframe plot method. 
# with index as the x axis
data.plot()
```

```python
# Create subsets of the original data. 
# Using index values that are less than 01/01/2006
data_recent = data[data.index > pd.datetime(year=2006, month=1, day=1)]
```

```python
data_recent.plot()
```

```python
# Scatter plot with oil prices on the x axis and price of gold on the y axis.
plt.scatter(data_recent.OIL, data_recent.GOLD, s=2)
```

```python
# To customize the plot.  First create a subplot as usual. 
# Then pass it to the dataframe's plot() method's ax argument. 
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set_ylabel("USD")

data_recent.plot(ax=ax)
```

```python


fig = plt.figure(figsize=(8,6))

ax1 = fig.add_subplot(211)
ax1.set_title("Oil")
data_recent.OIL.plot(ax=ax1)

ax2 = fig.add_subplot(212)
ax2.set_title("Gold")
data_recent.GOLD.plot(ax=ax2)

fig.tight_layout() # Adjust the layout to prevent overlaps

```

```python
# To display data on different scales is using a secondary y axis. Use 
# the secondary_y=True. 
# This will produce two y axis. 
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)

ax.set_ylabel("Gold (USD)")

data_recent.GOLD.plot(ax=ax, c='#EDDA74', linewidth=2)
ax2 = data_recent.OIL.plot(ax=ax, secondary_y=True, c='k') # Create the secondary y axis
ax2.set_ylabel("Oil (USD)")
```

```python
# Simple method to just plot the whole data frame. 
# Pass a list of column names to plot on the secondary axis to the 
# secondary_y argument. 
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)

data_recent.plot(ax=ax, secondary_y=["OIL"], color=["#EDDA74", "k"])
```
