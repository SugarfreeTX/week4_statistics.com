```python
def make_date(date_str):
    m, d, y = date_str.split("/")
    m =int(m)
    d=int(d)
    y=int(y)
    if y > 13:
        y += 1900
    else:
        y += 2000
    return dt.datetime(y, m, d)
```

# a) Create a scatter-plot with Movie Budgets as the X axis, and Profit Margin (Total Revenue / Budget) as the Y axis. Resize the Y axis as needed to convey the most information. (5 points)

```python
# worldwide gross / budget = profit margin
movies["Profit_Margin"] = movies["Worldwide Gross"] / movies["Budget"]
```

```python
movies.head()
```

```python
df = movies[["Budget", "Profit_Margin"]]
```

```python
df.head()
```

```python
x_data = df["Profit_Margin"]
y_data = df["Budget"]
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)

ax.set_xlabel("Movie Budget")
ax.set_ylabel("Profit Margin")
ax.scatter(x_data, y_data, s=32)
```

```python
x3_data = df3["Profit_Margin"]
y3_data = df3["Budget"]

colors = []
for y in y3_data:
    if y < 100000:colors.append('#0000FF')
    else: colors.append('#7FFFD4')

fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(111)
ax.set_xlabel("Budget")
ax.set_ylabel("Profit Margin")
ax.scatter(x3_data, y3_data, s=10, c=colors)
ax.set_xlim(-1000, 15000)
#ax.set_ylim(-1, 4.0)
```
